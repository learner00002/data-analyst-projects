### FEATURE ENGINEERING 

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['month'] = data['Order Date'].apply(lambda x:x.month)
```

```python
data.to_csv("cleaned_superstore.csv", index=False)
```

## Exploratory Data Analysis

## Sales by Category

```python
sales_by_category = data.groupby('Category')['Sales'].sum().sort_values()
sales_by_category.plot(kind='bar', title='Total Sales by Category')
plt.ylabel('Sales')
plt.show()
```

## Profit by Region

```python
profit_by_region = data.groupby('Region')['Profit'].sum().sort_values()
profit_by_region.plot(kind='barh', title='Profit by Region')
plt.xlabel('Profit')
plt.show()
```

##  Monthly Sales Trend

```python
monthly_sales = data.groupby('month')['Sales'].sum()
monthly_sales.plot(kind='line', marker='o', title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
